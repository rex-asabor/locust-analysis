```python
import cv2
from matplotlib import pyplot as plt
import random, os
import numpy as np
from PIL import Image
```

```python
img = Image.open('/Users/rexasabor/Downloads/Data/manual/downsampled_133_10sec_1910_1920_manual.tif')
img.seek(1)
plt.imshow(img, cmap="gray", vmin=0, vmax=256)
plt.show()
```

```python
open_cv_image = np.array(img) 
# Convert RGB to BGR 
gray_filtered = cv2.inRange(open_cv_image, 0, 130)
plt.imshow(gray_filtered, cmap="gray", vmin=0, vmax=256)
plt.show()
```
